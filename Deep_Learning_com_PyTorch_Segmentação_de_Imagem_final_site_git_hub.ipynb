# Carregamento da Imagem e Normalização

Na Segmentação de Imagem a máscara é carregada como uma imagem em tons de cinza e normalizada para uso em tarefas de segmentação de imagem, onde cada pixel representa a presença ou ausência de um objeto.   


### Explicando:

1. **`row = df.iloc[8]`**:
   - `df` é o DataFrame que contém os dados carregados do arquivo CSV.
   - `iloc[8]` é um método do Pandas usado para acessar uma linha específica pelo seu índice numérico (neste caso, a quinta linha, pois os índices começam em 0).

2. **Extração de Caminhos de Arquivo**:
   - `image_path = row["images"]`: Extrai o valor da coluna "images" da linha selecionada (`row`), que contém o caminho para a imagem.
   - `mask_path = row["masks"]`: Extrai o valor da coluna "masks" da mesma linha (`row`), que contém o caminho para a máscara associada à imagem.

3. **Carregamento e Manipulação de Imagem**:
   - `image = cv2.imread(image_path)`: Usa a função `cv2.imread()` do OpenCV para ler a imagem do caminho especificado (`image_path`).
   - `image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)`: Converte a imagem do formato padrão BGR (usado pelo OpenCV) para o formato RGB, que é mais comumente usado em bibliotecas como Matplotlib e NumPy para manipulação de imagens.

4. **Carregamento da Máscara e Normalização**:
   - `mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE) / 255.0`: Carrega a máscara como uma imagem em tons de cinza usando `cv2.imread()` com a flag `cv2.IMREAD_GRAYSCALE`. Em seguida, divide por 255.0 para normalizar os valores dos pixels para o intervalo entre 0 e 1, o que é comum em tarefas de segmentação onde a máscara binária deve ter valores de pixel de 0 (fundo) ou 1 (objeto).


### Motivo da divisão por 255.0

Essa divisão ocorre para a normalização de imagens.
Quando uma imagem em escala de cinza é carregada usando cv2.imread() com a flag cv2.IMREAD_GRAYSCALE, cada pixel da imagem é representado por um valor inteiro entre 0 e 255. Neste intervalo:

**0 representa o valor mínimo (preto).**

**255 representa o valor máximo (branco)**

### Razões para Normalização:

 - Compatibilidade com Modelos de Aprendizado de Máquina: Muitos algoritmos de aprendizado de máquina e redes neurais preferem trabalhar com dados normalizados, ou seja, valores de entrada que estão na mesma escala. Normalizar para o intervalo [0, 1] é uma prática comum para garantir que os modelos de machine learning converjam mais rapidamente e de maneira mais estável.

- Funções de Ativação: Alguns tipos de funções de ativação, como a função sigmóide, são mais eficazes quando os valores de entrada estão no intervalo [0, 1]. Isso pode ajudar na saída do modelo em problemas de classificação binária, por exemplo.

- Redução de Efeito de Diferenças de Escala: Normalizando para [0, 1], você reduz o impacto que grandes variações nos valores dos pixels podem ter sobre o processo de otimização durante o treinamento do modelo.


````python
row = df.iloc[8]
image_path = row["images"]
mask_path = row["masks"]

image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE) / 255.0

````

Essas transformações incluem rotação, escala, mudanças de contraste, recortes aleatórios, entre outras.

O parâmetro `is_check_shapes=False` na função get_train_augs() especifica se as transformações realizadas pelas Augmentations do albumentations devem verificar se as formas dos dados de entrada e saída são consistentes durante a composição das transformações.

````python
import albumentations as A

def show_image(image, mask):
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(np.transpose(image.numpy(), (1, 2, 0)))
    ax[0].set_title('Image')
    ax[1].imshow(mask.numpy()[0], cmap='gray')
    ax[1].set_title('Mask')
    plt.show()

def get_train_augs():
    return A.Compose([
        A.Resize(IMAGE_SIZE, IMAGE_SIZE),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
    ], is_check_shapes=False)

def get_valid_augs():
    return A.Compose([
        A.Resize(IMAGE_SIZE, IMAGE_SIZE)
    ], is_check_shapes=False)
````

  - `np.transpose(image, (2, 0, 1))`: Transpõe a imagem para o formato `(C, H, W)` (canal de cores, altura, largura).
  - `np.transpose(mask, (2, 0, 1))`: Transpõe a máscara para o formato `(C, H, W)`.
  - `image = torch.Tensor(image) / 255.0`: Converte a imagem para um tensor do PyTorch e normaliza os valores dos pixels para o intervalo [0, 1].
  - `mask = torch.round(torch.Tensor(mask) / 255.0)`: Converte a máscara para um tensor do PyTorch, normaliza os valores dos pixels para o intervalo [0, 1] e arredonda para garantir que os valores sejam binários (0 ou 1).


- Retorna a imagem e máscara preparadas como tensores do PyTorch para serem utilizadas no treinamento do modelo.


````python
from torch.utils.data import Dataset

class SegmentationDataset(Dataset):
    def __init__(self, df, augmentations):
        self.df = df
        self.augmentations = augmentations

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        image_path = row.images
        mask_path = row.masks

        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        mask = np.expand_dims(mask, axis=-1)


        if self.augmentations:
            data = self.augmentations(image=image, mask=mask)
            image = data['image']
            mask = data['mask']

        # Transpose image and mask to (C, H, W)
        image = np.transpose(image, (2, 0, 1)).astype(np.float32)
        mask = np.transpose(mask, (2, 0, 1)).astype(np.float32)
        image = torch.Tensor(image) / 255.0
        mask = torch.round(torch.Tensor(mask) / 255.0)

        return image, mask
````

# Definição do Modelo de Segmentação

Esta fase envolve a definição de um modelo de segmentação de imagens usando a arquitetura Unet da biblioteca `segmentation_models_pytorch`. O modelo é configurado com um inicializador (`__init__`) que define seus componentes principais e um método `forward` que define como os dados fluem pelo modelo durante a inferência e o treinamento. O uso de funções de perda como `DiceLoss` e `BCEWithLogitsLoss` dentro do método `forward` é fundamental para calcular a perda durante o treinamento, ajudando o modelo a aprender a segmentar corretamente as imagens.

### Explicando:

```python
class SegmentationModel(nn.Module):
    def __init__(self):
        super(SegmentationModel, self).__init__()
        self.arc = smp.Unet(
            encoder_name=ENCODER,
            encoder_weights=WEIGHTS,
            in_channels=3,
            classes=1,
            activation=None
        )

    def forward(self, images, masks=None):
        logits = self.arc(images)
        if masks is not None:
            loss1 = DiceLoss(mode='binary')(logits, masks)
            loss2 = nn.BCEWithLogitsLoss()(logits, masks)
            return logits, loss1 + loss2
        return logits
```

1. **Definição da Classe `SegmentationModel`**:
   - `class SegmentationModel(nn.Module):` Define uma nova classe `SegmentationModel` que herda funcionalidades da classe `nn.Module` do PyTorch.

Esta classe serve como o modelo principal para realizar a segmentação de imagens.


2. **Método `__init__` (Inicialização)**:
   - `super(SegmentationModel, self).__init__()` Chama o inicializador da classe `nn.Module` para inicializar a classe base.
   - `self.arc = smp.Unet(...)` Define o atributo `arc` como uma instância do modelo `Unet` da biblioteca `segmentation_models_pytorch`.
     - **`smp.Unet`**: Cria uma arquitetura de rede Unet para segmentação de imagens.
     - **Parâmetros do `Unet`**:
       - `encoder_name=ENCODER`: Define o nome do codificador (encoder) no código como `'timm-efficientbet-b0'`, o qual foi definido anteriormente em: Configurações de Inicialização: Comandos e Variáveis.
       - `encoder_weights=WEIGHTS`: Define os pesos do encoder, possivelmente treinados `'imagenet'`.
       - `in_channels=3`: Número de canais de entrada, geralmente 3 para imagens coloridas RGB.
       - `classes=1`: Número de classes de saída, aqui configurado para 1 pois é um problema de segmentação binária.
       - `activation=None`: A função de ativação será aplicada posteriormente, após o cálculo dos logits.

3. **Método `forward`**:
   - `def forward(self, images, masks=None):` Define o método `forward`, que processa a passagem direta (forward pass) dos dados pelo modelo.
     - **`images`**: Entrada de imagens para o modelo.
     - **`masks`**: Máscaras (rótulos) correspondentes às imagens, opcionalmente fornecidas durante o treinamento para calcular a perda.

   - `logits = self.arc(images)`: Realiza a passagem direta das imagens através do modelo `Unet` para obter os logits (saídas do modelo antes da aplicação da função de ativação).

   - `if masks is not None:`: Verifica se as máscaras foram fornecidas.
     - `loss1 = DiceLoss(mode='binary')(logits, masks)`: Calcula a perda usando a função de perda `DiceLoss` para avaliar a sobreposição entre as predições (`logits`) e as máscaras reais (`masks`).
     - `loss2 = nn.BCEWithLogitsLoss()(logits, masks)`: Calcula a perda usando a função de perda `BCEWithLogitsLoss`, que combina uma função sigmoide e a perda de entropia cruzada binária.

# Função para o Treinamento

### Função `train_fn`

```python
def train_fn(data_loader, model, optimizer):
    model.train()  # Coloca o modelo em modo de treinamento
    total_loss = 0.0  # Inicializa a variável para acumular a perda total

    for images, masks in data_loader:  # Itera sobre o data loader, fornecendo lotes de imagens e máscaras
        images = images.to(DEVICE)  # Move as imagens para o dispositivo (CPU ou GPU)
        masks = masks.to(DEVICE)  # Move as máscaras para o dispositivo (CPU ou GPU)

        optimizer.zero_grad()  # Zera os gradientes do otimizador

        logits, loss = model(images, masks)  # Faz uma previsão com o modelo e calcula a perda

        loss.backward()  # Calcula o gradiente da perda
        optimizer.step()  # Atualiza os parâmetros do modelo

        total_loss += loss.item()  # Acumula a perda total

    return total_loss / len(data_loader)  # Retorna a perda média
```

### Explicando:

1. **Definição da Função**
   ```python
   def train_fn(data_loader, model, optimizer):
   ```
   A função `train_fn` é definida para treinar o modelo. Ela recebe três argumentos:
   - `data_loader`: Um DataLoader PyTorch que fornece os dados de treinamento.
   - `model`: O modelo de segmentação a ser treinado.
   - `optimizer`: O otimizador usado para atualizar os pesos do modelo.

2. **Modo de Treinamento do Modelo**
   ```python
   model.train()
   ```
   Coloca o modelo em modo de treinamento. Isso ativa comportamentos específicos de treinamento, como dropout e batch normalization.

3. **Inicialização da Perda Total**
   ```python
   total_loss = 0.0
   ```
   Inicializa uma variável para acumular a perda total durante o treinamento.

4. **Iteração sobre o DataLoader**
   ```python
   for images, masks in data_loader:
   ```
   Itera sobre o `data_loader`, que fornece lotes de imagens (`images`) e suas respectivas máscaras (`masks`).

5. **Mover Dados para o Dispositivo**
   ```python
   images = images.to(DEVICE)
   masks = masks.to(DEVICE)
   ```
   Move as imagens e as máscaras para o dispositivo especificado (GPU no nosso caso, pois estamos usando CUDA). A variável `DEVICE` foi definida anteriormente no seu código. Para usar CPU seria `torch.device("cpu")`.

6. **Zerar os Gradientes**
   ```python
   optimizer.zero_grad()
   ```
   Zera os gradientes acumulados do otimizador. Isso é importante para evitar que os gradientes de diferentes lotes se acumulem.

7. **Previsão e Cálculo da Perda**
   ```python
   logits, loss = model(images, masks)
   ```
   Faz uma previsão com o modelo e calcula a perda. Presumivelmente, o modelo retorna tanto as previsões (`logits`) quanto a perda (`loss`) quando chamado com `images` e `masks` como argumentos.

8. **Retropropagação da Perda**
   ```python
   loss.backward()
   ```
   Calcula os gradientes da perda em relação aos parâmetros do modelo.

9. **Atualização dos Parâmetros do Modelo**
   ```python
   optimizer.step()
   ```
   Atualiza os pesos do modelo usando os gradientes calculados.

10. **Acumulação da Perda**
    ```python
    total_loss += loss.item()
    ```
    Converte a perda para um número Python usando `.item()` e acumula na variável `total_loss`.

11. **Retorno da Perda Média**
    ```python
    return total_loss / len(data_loader)
    ```
    Calcula e retorna a perda média dividindo a perda total pelo número de lotes no `data_loader`.

### Definição do Otimizador e da Função de Perda

Antes de usar a função `train_fn`, você precisa definir o otimizador e a função de perda:

```python
optimizer = torch.optim.Adam(model.parameters(), lr=LR)
```

- **Otimizador**: Aqui, o otimizador Adam é usado com a taxa de aprendizado (`lr`) definida por `LR`.







````python
def train_fn(data_loader, model, optimizer):
    model.train()  # Coloca o modelo no modo de treinamento
    total_loss = 0.0  # Inicializa a perda total

    for images, masks in data_loader:  # Itera sobre o trainloader
        images = images.to(DEVICE)  # Move as imagens para o dispositivo
        masks = masks.to(DEVICE)  # Move as máscaras para o dispositivo

        optimizer.zero_grad()  # Zera os gradientes do otimizador

        logits, loss = model(images, masks)  # Faz uma previsão com o modelo

        loss.backward()  # Calcula o gradiente da perda
        optimizer.step()  # Atualiza os parâmetros do modelo

        total_loss += loss.item()  # Acumula a perda total

    return total_loss / len(trainloader)  # Retorna a perda média

# Definição do otimizador e da função de perda
optimizer = torch.optim.Adam(model.parameters(), lr=LR)


````

# Avaliação
- `eval_fn` é comumente usada para avaliar o desempenho de um modelo após o treinamento em um conjunto de dados de validação ou teste.
- Ela calcula a perda média do modelo em relação aos dados fornecidos pelo `data_loader`.


```python
def eval_fn(data_loader, model):
    model.eval()  # Coloca o modelo no modo de avaliação
    total_loss = 0.0  # Inicializa a variável para acumular a perda total

    with torch.no_grad():  # Desabilita o cálculo automático de gradientes
        for images, masks in tqdm(data_loader):  # Itera sobre o data loader
            images = images.to(DEVICE)  # Move as imagens para o dispositivo (GPU ou CPU)
            masks = masks.to(DEVICE)  # Move as máscaras para o dispositivo (GPU ou CPU)

            logits, loss = model(images, masks)  # Faz uma previsão com o modelo e calcula a perda

            total_loss += loss.item()  # Acumula a perda total

    return total_loss / len(data_loader)  # Retorna a perda média
```

### Explicando:

1. **Colocar o Modelo em Modo de Avaliação**
   ```python
   model.eval()
   ```
   - `model.eval()` é usado para colocar o modelo no modo de avaliação. Isso altera o comportamento de certas camadas, como dropout e batch normalization, para que operem de maneira diferente durante a avaliação do que durante o treinamento.

2. **Inicialização da Variável de Perda Total**
   ```python
   total_loss = 0.0
   ```
   - `total_loss` é uma variável inicializada para acumular a perda total durante a avaliação.

3. **Desabilitar o Cálculo Automático de Gradientes**
   ```python
   with torch.no_grad():
   ```
   - `torch.no_grad()` é um contexto do PyTorch que desabilita o cálculo automático de gradientes. Isso melhora a eficiência computacional durante a avaliação, pois os gradientes não precisam ser rastreados ou armazenados.

4. **Iteração sobre o DataLoader**
   ```python
   for images, masks in tqdm(data_loader):
   ```
   - `data_loader` é iterado para obter lotes de dados (`images` e `masks`). `tqdm` é usado para criar uma barra de progresso durante a iteração, o que é útil para monitorar o progresso da avaliação.

5. **Mover Dados para o Dispositivo**
   ```python
   images = images.to(DEVICE)
   masks = masks.to(DEVICE)
   ```
   - `images.to(DEVICE)` e `masks.to(DEVICE)` movem os tensores de imagens e máscaras para o dispositivo especificado (`DEVICE`), no nosso caso, uma GPU.

6. **Fazer uma Previsão com o Modelo**
   ```python
   logits, loss = model(images, masks)
   ```
   - `model(images, masks)` faz uma previsão utilizando o modelo com as imagens de entrada (`images`) e as máscaras de saída esperadas (`masks`).
   - `logits` são as saídas não processadas do modelo antes da aplicação de uma função de ativação, geralmente usadas em problemas de classificação.
   - `loss` é a perda calculada entre as previsões (`logits`) e as máscaras de referência (`masks`).

7. **Acumular a Perda Total**
   ```python
   total_loss += loss.item()
   ```
   - `loss.item()` converte o tensor de perda para um valor Python e adiciona à `total_loss`.

8. **Retornar a Perda Média**
   ```python
   return total_loss / len(data_loader)
   ```
   - `total_loss / len(data_loader)` calcula a perda média dividindo a perda total pelo número de lotes no `data_loader`.
   - Retorna a perda média como resultado da função `eval_fn`.













````python
def eval_fn(data_loader, model):
  model.eval()
  total_loss = 0.0

  with torch.no_grad():
    for images, masks in tqdm(data_loader):

      images = images.to(DEVICE)
      masks = masks.to(DEVICE)

      logits, loss = model(images, masks)  # Faz uma previsão com o modelo

      total_loss += loss.item()  # Acumula a perda total

    return total_loss / len(trainloader)  # Retorna a perda média


````

# Loop de Treinamento do Modelo.

Este código implementa um loop de treinamento completo para um modelo de segmentação de imagens. Ele treina o modelo por várias épocas, avalia seu desempenho após cada época usando dados de validação e salva o melhor modelo encontrado com base na menor perda de validação observada. Esse processo é crucial para melhorar o desempenho do modelo ao longo do tempo, ajustando seus parâmetros para minimizar a perda durante o treinamento.

Vamos explicar passo a passo o código fornecido, que treina um modelo de rede neural em várias épocas (`EPOCHS`), salvando o melhor modelo com base na perda de validação (`valid_loss`):

```python
best_val_loss = float("inf")  # Inicializa o melhor valor de perda de validação como infinito positivo

for i in range(EPOCHS):  # Loop sobre o número de épocas especificado
    train_loss = train_fn(trainloader, model, optimizer)  # Treina o modelo no conjunto de treinamento
    valid_loss = eval_fn(validloader, model)  # Avalia o modelo no conjunto de validação

    if valid_loss < best_val_loss:  # Verifica se a perda de validação atual é menor que a melhor perda de validação registrada
        torch.save(model.state_dict(), 'best_model.pt')  # Salva os parâmetros do modelo atual como 'best_model.pt'
        print("SAVED-MODEL")  # Exibe uma mensagem indicando que o modelo foi salvo
        best_val_loss = valid_loss  # Atualiza o melhor valor de perda de validação com o valor atual

    print(f"Epoch: {i+1} Train loss: {train_loss} Valid loss: {valid_loss}")  # Imprime as perdas de treinamento e validação para a época atual
```

### Explicação Detalhada:

1. **Inicialização de `best_val_loss`**
   ```python
   best_val_loss = float("inf")
   ```
   - `best_val_loss` é inicializado com infinito positivo (`float("inf")`). Isso garante que o primeiro valor de `valid_loss` será considerado como o melhor inicialmente.

2. **Loop de Épocas**
   ```python
   for i in range(EPOCHS):
   ```
   - Itera sobre o número especificado de épocas (`EPOCHS`). Durante cada época, o modelo é treinado no conjunto de treinamento e avaliado no conjunto de validação.

3. **Treinamento do Modelo**
   ```python
   train_loss = train_fn(trainloader, model, optimizer)
   ```
   - `train_fn` é chamado para treinar o modelo (`model`) usando o otimizador (`optimizer`) e os dados de treinamento fornecidos pelo `trainloader`.
   - `train_loss` é a perda média durante o treinamento nesta época.

4. **Avaliação do Modelo**
   ```python
   valid_loss = eval_fn(validloader, model)
   ```
   - `eval_fn` é chamado para avaliar o modelo no conjunto de validação (`validloader`).
   - `valid_loss` é a perda média durante a avaliação no conjunto de validação nesta época.

5. **Atualização do Melhor Modelo**
   ```python
   if valid_loss < best_val_loss:
       torch.save(model.state_dict(), 'best_model.pt')
       best_val_loss = valid_loss
   ```
   - Verifica se a `valid_loss` atual é menor que `best_val_loss`.
   - Se sim, salva os parâmetros do modelo (`model.state_dict()`) como um arquivo `'best_model.pt'`.
   - Atualiza `best_val_loss` com o valor de `valid_loss`, indicando que encontramos um novo melhor modelo com base na perda de validação.

6. **Impressão das Métricas**
   ```python
   print(f"Epoch: {i+1} Train loss: {train_loss} Valid loss: {valid_loss}")
   ```
   - Imprime as perdas de treinamento (`train_loss`) e validação (`valid_loss`) para a época atual.
   - Fornece feedback sobre o desempenho do modelo após cada época de treinamento e avaliação.



````python
best_val_loss = float("inf")

for i in range(EPOCHS):
    train_loss = train_fn(trainloader, model, optimizer)
    valid_loss = eval_fn(validloader, model)  # Use validloader, model, e a função de perda

    if valid_loss < best_val_loss:
        torch.save(model.state_dict(), 'best_model.pt')
        print("SAVED-MODEL")
        best_val_loss = valid_loss  # Atualiza o melhor valor de validação

    print(f"Epoch: {i+1} Train loss: {train_loss} Valid loss: {valid_loss}")

````

# Inferência do Modelo ou Predição do Modelo
Nesta fase, o modelo treinado é utilizado para realizar predições em uma imagem específica do conjunto de validação. A inferência produz uma máscara binária que indica a presença ou ausência de objetos de interesse na imagem, com base nos parâmetros aprendidos pelo modelo durante o treinamento.

Este código é utilizado após o treinamento do modelo para fazer previsões em novas imagens do conjunto de validação (`validset`).
- Ele carrega o modelo treinado que teve o melhor desempenho de validação (`best_model.pt`), aplica-o a uma imagem específica (`validset[idx]`) e gera uma máscara binária que representa a segmentação prevista pelo modelo.


### Explicando:

1. **Carregamento do Melhor Modelo**
   ```python
   model.load_state_dict(torch.load('/content/best_model.pt'))
   ```
   - `torch.load('/content/best_model.pt')` carrega os parâmetros salvos do modelo do arquivo `'best_model.pt'`.
   - `model.load_state_dict(...)` carrega esses parâmetros para o modelo `model`, restaurando-o para o estado em que era considerado o melhor em termos de perda de validação durante o treinamento.

2. **Obtenção da Amostra de Validação**
   ```python
   image, mask = validset[idx]
   ```
   - `validset[idx]` retorna a imagem e a máscara correspondente no índice `idx` do conjunto de validação (`validset`).

3. **Previsão com o Modelo**
   ```python
   logits = model(image.to(DEVICE).unsqueeze(0))
   ```
   - `image.to(DEVICE)` move a imagem para o dispositivo especificado (`DEVICE`), é uma uma GPU.
   - `.unsqueeze(0)` adiciona uma dimensão extra à imagem para se adequar à entrada esperada pelo modelo. A dimensão adicionada é para o tamanho do lote (batch size), que neste caso é 1.
   - `model(image.to(DEVICE).unsqueeze(0))` passa a imagem através do modelo para gerar previsões (`logits`). `logits` são as saídas do modelo antes da aplicação da função de ativação final.

4. **Aplicação da Função Sigmoid**
   ```python
   pred_mask = torch.sigmoid(logits)
   ```
   - `torch.sigmoid(logits)` aplica a função sigmoid aos `logits`, transformando-os em probabilidades.

5. **Conversão para Máscara Binária**
   ```python
   pred_mask = (pred_mask > 0.5) * 1.0
   ```
   - `(pred_mask > 0.5)` cria uma máscara booleana onde os valores maiores que 0.5 são True (1) e os valores menores ou iguais a 0.5 são False (0).
   - `* 1.0` multiplica a máscara booleana por 1.0 para converter os True e False em valores 1.0 e 0.0, respectivamente, resultando na máscara binária final.





````python
idx = 35
model.load_state_dict(torch.load('/content/best_model.pt'))

image, mask = validset[idx]

logits = model(image.to(DEVICE).unsqueeze(0))  # (C, H, W) -> (1, C, H, W)
pred_mask = torch.sigmoid(logits)
pred_mask = (pred_mask > 0.5) * 1.0



````
